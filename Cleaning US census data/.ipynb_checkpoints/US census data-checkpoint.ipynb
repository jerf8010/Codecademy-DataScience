# Cleaning US Census Data


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
print(pd.read_csv('state0.csv'))
print(pd.read_csv('state1.csv'))
print(pd.read_csv('state2.csv'))
print(pd.read_csv('state3.csv'))
print(pd.read_csv('state4.csv'))
print(pd.read_csv('state5.csv'))
print(pd.read_csv('state6.csv'))
print(pd.read_csv('state7.csv'))
print(pd.read_csv('state8.csv'))
```

### Concatenating files and converting to DF

```python
import glob

files = glob.glob('state*.csv')
lista= []

for filename in files:
    data = pd.read_csv(filename)
    lista.append(data)

us_census = pd.concat(lista).reset_index()
print(us_census)
    
```

### Columns dtypes

```python
print(us_census.columns)
```

```python
print(us_census.dtypes)
```

### Transforming Income column to numeric

```python
us_census.Income = us_census.Income.str.replace('$','', regex = True)
us_census.Income = pd.to_numeric(us_census.Income)
print(us_census.Income)
print(us_census.dtypes)
```

### Gender column

```python
gender = us_census.GenderPop.str.split("_")
male = gender.str.get(0)
female = gender.str.get(1)

male = male.str.replace('M', '')
female = female.str.replace('F','')

male = pd.to_numeric(male)
female = pd.to_numeric(female)

us_census['male_float'] = male
us_census['female_float'] = female

print(us_census.dtypes)
```

### Visualizing wiht NAN

```python
plt.scatter(us_census.female_float, us_census.male_float)
plt.show()
```

### Correcting NAN values for male and female

```python
us_census = us_census.fillna(value = {'female_float' : us_census.TotalPop - us_census.male_float, 'male_float': us_census.TotalPop-us_census.female_float})

print(us_census.female_float)
print(us_census.male_float)
```

### Visualizing without NAN

```python
plt.scatter(us_census.female_float, us_census.male_float)
plt.show()
```

### Looking for duplicated rows

```python
dupl = us_census.duplicated()
print(dupl.value_counts())

us_census.drop_duplicates()
dupl1 = us_census.duplicated()
print(dupl1.value_counts())
```

## About the race

```python
us_census.Hispanic = us_census.Hispanic.str.replace('%','')
us_census.Hispanic = pd.to_numeric(us_census.Hispanic)
us_census = us_census.fillna(value = {'Hispanic' : 0})

us_census.White = us_census.White.str.replace('%','')
us_census.White = pd.to_numeric(us_census.White)
us_census = us_census.fillna(value = {'White' : 0})

us_census.Black = us_census.Black.str.replace('%','')
us_census.Black = pd.to_numeric(us_census.Black)
us_census = us_census.fillna(value = {'Black' : 0})

us_census.Native = us_census.Native.str.replace('%','')
us_census.Native = pd.to_numeric(us_census.Native)
us_census = us_census.fillna(value = {'Native' : 0})

us_census.Asian = us_census.Asian.str.replace('%','')
us_census.Asian = pd.to_numeric(us_census.Asian)
us_census = us_census.fillna(value = {'Asian' : 0})

us_census.Pacific = us_census.Pacific.str.replace('%','')
us_census.Pacific = pd.to_numeric(us_census.Pacific)
us_census = us_census.fillna(value = {'Pacific' : 0})
```
